```python
import pandas as pd
import sns as sns

# Replace 'data.csv' with your actual CSV file path if it's in a different location
csv_file_path = 'graf.csv'

# Use the read_csv() function to read the CSV file and store it in a DataFrame
df = pd.read_csv(csv_file_path, sep=';')

df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

ax = plt.gca()
ax.set_title("Highest fitness in each generation")
ax.set_xlabel("Generations")
ax.set_ylabel("Fitness")

df.rename(columns={'x': 'Generations', 'y': 'Fitness', 'z':'Curriculum levels'}, inplace=True)
sns.(x='Generations', y='Fitness' , hue='Curriculum levels', data=df, style='Curriculum levels').figure.savefig("fitness-over-generations.pdf")
# Name the legend values
plt.show()
```

```python
df.groupby('Curriculum levels').mean()[['Fitness']]
```

```python
df.groupby('Curriculum levels').median()[['Fitness']]
```

```python
df.groupby('Curriculum levels').max()[['Fitness']] - df.groupby('Curriculum levels').min()[['Fitness']]
```

```python
df_zero = pd.read_csv("zero_test_mse_results.csv", sep=';')
df_zero['Curriculum levels'] = "zero"
df_two = pd.read_csv("two_test_mse_results.csv", sep=';')
df_two['Curriculum levels'] = "two"
df_three = pd.read_csv("three_test_mse_results.csv", sep=';')
df_three['Curriculum levels'] = "three"

df_all = pd.concat([df_zero, df_two, df_three])
sns.boxenplot(data=df_all, x="MSE", y="Curriculum levels").figure.savefig("boxenplot.pdf")
```
